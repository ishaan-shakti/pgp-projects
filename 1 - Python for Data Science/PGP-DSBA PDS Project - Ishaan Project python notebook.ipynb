```python
df['Gender'] = df['Gender'].replace(['Femal','Femle'],'Female') #Code to replace words spelled incorrectly with the correct word
df['Gender'].unique() #Check to ensure the mistakes have been corrected
```

```python
#Treatment of null values in the dataframe
df.loc[df['Gender'].isnull()==True]
```

```python
#Code to impute null values in 'gender' column with the mode
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
```

```python
#Check if null values have been replaced sucessfully
df['Gender'].value_counts(dropna=False)
```

```python
#Analysis of null values in 'Partner_salary'
df.loc[df['Partner_salary'].isnull()==True]
```

```python
#Imputing null values in 'Partner_salary' based on the rule that 'Total_salary'= 'Salary'+'Partner_Salary'
#so 'Parter_salary'='Total_salary'-'Salary' if 'Partner_working is 'Yes" otherwise the value is 0
df.loc[df['Partner_working'] == 'No', 'Partner_salary'] = df['Partner_salary'].fillna(0)
df.loc[df['Partner_working'] == 'Yes', 'Partner_salary'] = df['Total_salary'] - df['Salary']
```

```python
#Converting 'Partner_salary' to integer data type
df['Partner_salary'] = df['Partner_salary'].astype('int64')
```

```python
#Recheck of null values
df.isnull().sum()
```

### **Key Questions**

```python
#1. Do men tend to prefer SUVs more compared to women?
plt.figure(figsize=(10,7))
sns.countplot(df, x="Make", hue="Gender")
plt.show()
```

```python
#2. What is the likelihood of a salaried person buying a Sedan?
plt.figure(figsize=(10, 7))
sns.countplot(data=df, x="Profession", hue="Make")
plt.show()
```

```python
#3. What evidence or data supports Sheldon Cooper's claim that a salaried male is an easier target for a SUV sale over a Sedan sale?
plt.figure(figsize=(10, 7))
sns.countplot(data=df[df['Gender']=='Male'], x="Profession", hue="Make")
plt.show()
```

```python
#4. How does the average amount spent on purchasing automobiles vary by gender?
plt.figure(figsize=(6, 4))
sns.barplot(data=df, x="Gender", y="Price", hue="Gender")
plt.show()
```

```python
#Mean values of total spent by gender
df.groupby('Gender')['Price'].mean()
```

```python
#5. How much money was spent on purchasing automobiles by individuals who took a personal loan?
plt.figure(figsize=(6, 4))
sns.barplot(data=df, x="Personal_loan", y="Price",hue="Personal_loan")
plt.show()
```

```python
#average sum of money spent by individuals who took out a personal loan
df.groupby('Personal_loan')['Price'].mean()
```

```python
#6. How does having a working partner influence the purchase of higher-priced cars?
plt.figure(figsize=(6, 4))
sns.barplot(data=df[df['Price']>47000], x="Partner_working", y="Price", hue="Partner_working")
plt.show()
```

```python
#average cost of high priced car by working partners
df[df['Price']>47000].groupby('Partner_working')['Price'].mean()
```

```python
#median cost of high priced car by working partners
df[df['Price']>47000].groupby('Partner_working')['Price'].median()
```

```python
#analysing the mode of "make" for each "gender" group and sub grouped by "marital_status"
df.groupby(['Gender', 'Marital_status'])[['Make']].apply(lambda x: x.mode().iloc[0])
```
